```python

n = 4
val = [['.' for i in range(n)] for j in range(n)]

val[0][1] = 'Q'

for i in range(n):
    print("+" + "---+" * n)  
    for j in range(n):
        print("|", end=' ')
        print(val[i][j], end=' ')  
    print("|")
print("+" + "---+" * n)  
```
